## Agente Reativo Simples: Robô Aspirador de Pó

```python
# Bibliotecas necessarias

import time
import random
from random import randint
```

```python
# Definindo variaveis

agente = 'A'

limpo = 0

```

```python
opcoes = [0,1]
ambiente = [[random.choice(opcoes) for _ in range(3)] for _ in range(3)]
ambiente[0][0] = "A"

print(ambiente)
```

```python
def limpo(ambiente):
    for linha in ambiente:
        for sala in linha:
            if sala == 1:
                return False
    return True
```

```python
def mover_agente(direcao, linha, coluna):
    if direcao == "cima" and linha > 0:
        linha -= 1
    elif direcao == "baixo" and linha < 2:
        linha += 1
    elif direcao == "esquerda" and coluna > 0:
        coluna -= 1
    elif direcao == "direita" and coluna < 2:
        coluna += 1
    else:
        print("Movimento inválido ou limite atingido.")
        
    return linha, coluna
```

```python
def mostrar_na_tela(ambiente):
    for linha in ambiente:
        print(linha)
    print() 
```

```python
def aspirador(ambiente):
    v_limpo = 0 
    linha = 0
    atual = 0 
    
    while(True):
        if ambiente[linha][atual] != v_limpo:
                
            ambiente[linha][atual] = agente
            mostrar_na_tela(ambiente)
            time.sleep(0.5)
            ambiente[linha][atual] = v_limpo
            
            if atual == 0 and (atual + 1 <= 2 and ambiente[linha][atual + 1] != v_limpo):
                atual = atual + 1
                    
            elif atual == 0 and (atual + 1 <= 2 and ambiente[linha][atual + 1] == v_limpo) and (atual + 2 > 2 or ambiente[linha][atual + 2] == v_limpo):
                if linha < 2:
                    linha = linha + 1
                    atual = 0
                else:
                    print("Todas as salas estão limpas")
                    break
                    
            elif atual == 0 and (atual + 1 <= 2 and ambiente[linha][atual + 1] == v_limpo) and (atual + 2 <= 2 and ambiente[linha][atual + 2] != v_limpo):
                atual = atual + 1    
                    
            elif atual == 1 and ambiente[linha][atual - 1] != v_limpo:
                atual = atual - 1
                    
            elif atual == 1 and (atual + 1 <= 2 and ambiente[linha][atual + 1] != v_limpo):
                atual = atual + 1
                
            elif atual == 1 and ambiente[linha][atual - 1] == v_limpo and (atual + 1 <= 2 and ambiente[linha][atual + 1] == v_limpo):
                if linha < 2:
                    linha = linha + 1
                    atual = 0
                else:
                    print("Todas as salas estão limpas")
                    break
                    
            elif atual == 2 and ambiente[linha][atual - 1] != v_limpo:
                atual = atual - 1
                    
            elif atual == 2 and ambiente[linha][atual - 1] == v_limpo:
                if linha < 2:
                    linha = linha + 1
                    atual = 0
                else:
                    print("Todas as salas estão limpas")
                    break
        else:
            salva_estado = ambiente[linha][atual]
            ambiente[linha][atual] = agente 
            mostrar_na_tela(ambiente)
            time.sleep(0.5)
            
            ambiente[linha][atual] = salva_estado

            if atual == 0 and (atual + 1 <= 2 and ambiente[linha][atual + 1] != v_limpo):
                atual = atual + 1
            elif atual == 0 and (atual + 1 <= 2 and ambiente[linha][atual + 1] == v_limpo) and (atual + 2 > 2 or ambiente[linha][atual + 2] == v_limpo):
                if linha < 2:
                    linha = linha + 1
                    atual = 0
                else:
                    print("Todas as salas estão limpas")
                    break
            elif atual == 0 and (atual + 1 <= 2 and ambiente[linha][atual + 1] == v_limpo) and (atual + 2 <= 2 and ambiente[linha][atual + 2] != v_limpo):
                atual = atual + 1    
            elif atual == 1 and ambiente[linha][atual - 1] != v_limpo:
                atual = atual - 1
            elif atual == 1 and (atual + 1 <= 2 and ambiente[linha][atual + 1] != v_limpo):
                atual = atual + 1
            elif atual == 1 and ambiente[linha][atual - 1] == v_limpo and (atual + 1 <= 2 and ambiente[linha][atual + 1] == v_limpo):
                if linha < 2:
                    linha = linha + 1
                    atual = 0
                else:
                    print("Todas as salas estão limpas")
                    break
            elif atual == 2 and ambiente[linha][atual - 1] != v_limpo:
                atual = atual - 1
            elif atual == 2 and ambiente[linha][atual - 1] == v_limpo:
                if linha < 2:
                    linha = linha + 1
                    atual = 0
                else:
                    print("Todas as salas estão limpas")
                    break
                
            continue
                
    return ambiente
```
